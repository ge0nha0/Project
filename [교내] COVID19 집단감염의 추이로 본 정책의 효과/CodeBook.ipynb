## 데이터 설명

### <font color='scarlet'>Main Data</font>

#### <font color='scarlet'>01. PatientInfo.csv</font>
2020.01.20-2020.06.30 까지의 코로나 환자 데이터
- patient_id : 환자 ID   
- sex : 환자의 성별   
- age : 환자의 연령층  
- country : 환자의 국적   
- province : 감염 주소지(시)
- city : 감염 주소지(도)
- infection_case : 감염 경로   
- infected_by : 누구로 인해 감염되었는지(감염시킨 환자 번호)  
- contact_number : 접촉자 수  
- symptom_onset_date : 증상 발생일  
- confirm_date : 확진 판정일  
- relased_date : 격리 해제일  
- deceased_date : 사망일
- state : 현재 상태

#### <font color='scarlet'>02. Region.csv</font>
지역 정보 데이터
- code : 지역 코드   
- proince : 지역(시도)
- city : 지역(시군)
- latitude : 위도   
- longitude : 경도  
- elementary_school_count : 초등학교 수  
- kindergarten_count : 유치원 수  
- university_count : 대학교 수  
- academy_ratio : 학원 비율  
- elderly_population_ratio : 노인 인구 비율  
- elderly_alone_ratio : 독거노인 비율   
- nursing_home_count : 요양원 수

#### <font color='scarlet'>03. Weather.csv</font>
날씨 데이터
- code : 날씨 코드   
- province : 지역   
- date : 날짜   
- ave_temp : 평균 기온   
- min_temp : 최저 기온   
- max_temp : 최고 기온   
- precipitation : 강수량        
- max_wind_speed : 최고 풍속   
- most_wind_direction : 주된 바람 방향    
- avg_relative_humidity : 평균 상대습도   

#### <font color='scarlet'>04. Searchtrend.csv</font>
2016.01.01-2022.06.29 까지의 네이버에서 해당 단어 검색 비율 데이터
- date : 데이터 집계 날짜   
- cold : '감기' 검색 비율    
- flu : '독감' 검색 비율   
- pneumonia : '폐렴' 검색 비율    
- coronavirus : '코로나' 검색 비율

### <font color='coral'>Sub Data</font>

#### <font color='coral'>05. Policy.csv</font>
정부가 날짜별로 어떠한 코로나 관련 정책을 발령했는지에 대한 데이터
- policy_id : 정책 ID
- type : 정책의 유형   
- gov_policy : 정책명   
- detail : 정책 세부사항
- start_date : 정책 시작일  
- end_date : 정책 종료일  

#### <font color='coral'>06. Case.csv</font>
유명한 코로나 발생 케이스에 대한 데이터
- case_id : case ID (첫 5자리는 region의 code를 나타냄)   
- proince : 지역(시도)
- city : 지역(시군)
- group : 집단 감염 여부 (True : 집단감염, contact with patient : 확진자 접촉, overseas inflow : 해외유입, etc : 기타)       
- infection_case : 해외 감염 (etc : 우선 개별 감염이지만, 집단 감염일 가능성이 커 아직 역학조사 진행중인 상황)  
- confirmed : 감염 환자 수
- latitude : 위도   
- longitude : 경도
