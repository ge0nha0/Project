```python
import pandas as pd
```

```python
df1 = pd.read_csv('clothes_info_긴팔_006001003_page17_20230630.csv')
df2 = pd.read_csv('clothes_info_모자_006004003_page12_20230630.csv')
df3 = pd.read_csv('clothes_info_바지_006001006_page17_20230630.csv')
df4 = pd.read_csv('clothes_info_반팔_006001002_page17_20230630.csv')
```

```python
# 이미지 URL 목록
urls = df3['img_url'].tolist() # df2 # df3 # df4
```

```python
import os
import time
from PIL import Image

# Create the "hoodie" folder if it doesn't exist
if not os.path.exists("pants"):
    os.makedirs("pants")

# Image download function
def download_image(url, filename):
    os.system(f"curl {url} > pants/{filename}")

# Download and return images
def fetch_images(urls):
    images = []
    for url in urls:
        filename = url.split("/")[-1] # Extract image filename
        download_image(url, filename) # Download image
        img = Image.open(f"pants/{filename}") # Open image
        images.append(img) # Add to image list
    return images

# Execution example
start = time.time() # Start checking time
downloaded_images = fetch_images(urls) # Download and return images
print("Image download time:", time.time() - start)

# Check the downloaded image
for i, img in enumerate(downloaded_images):
    img.show() # View image
```
