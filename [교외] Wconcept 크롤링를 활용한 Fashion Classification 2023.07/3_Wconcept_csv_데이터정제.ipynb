#### material 정제하기

```python
replace_dic = {"COTTON": "면", "코튼":"면", "POLYESTER":"폴리에스터", "MODAL":"모달", "SPAN":"스판", "MAIN-RAYON":"메인레이온",
               "POLYAMIDE":"폴리아미드", "NYLON":"나일론", "TENCEL":"텐셀", "WOOL":"울", "ACRYLIC":"아크릴", "ELASTANE":"엘라스테인",
               "POLYURETHANE":"폴리우레탄", "CASHMERE":"캐시미어", "RAYON":"레이온", "POLY":"폴리", "VISCOSE":"인견", "ANGORA":"앙고라",
               "CHIEF":"겉감", "RIB":"안감", "SPANDEX":"스판덱스", "MAIN RAYON": "메인레이온", "RIBBED":"겉감", "SLUB":"안감", "덱스":"스판",
               "FRONT":"겉감", "BACK":"안감", "몸판":"겉감", "배색":"안감", "POLY.U":"폴리우레탄", "SPANDEX":"스판", "DEX":"스판", "SUB":"안감",
               "PU": "폴리우레탄", "LINING-VISCOSE":"라이닝-비스코스", "LINEN":"린넨", "YOCELL":"요셀", "SHELL":"겉감", "LINING":"안감", "METAL":"메탈"}
```

```python
# raplace
material = df["clothes_material"]
new_material = []
to_replace = [":", " ", "(", ")", "%", "-", '"', "/", "[", "]", "+", "겉감", "안감", "\n", ".", "*", ",", "<", ">"
         "FRONT", "BACK", "몸판", "배색", "SHELL", "OUT", "MAIN", "BODY", "<Main fabric>", "<Sub fabric>"] # 해당 단어는 빈칸으로 대체함.

for text in material:

    for re in to_replace:
        text = text.replace(re, " ") #to_replace들을 빈칸으로 없앤다.
    for key, value in replace_dic.items():
        if key in text:
            text = text.replace(key, f" {value} ") # replace_dic의 key를 value로 바꿔주고 공백을 넣어준다. 코튼100과 같은 걸 바꿔줄려고

        if value in text:
            text = text.replace(value, f" {value} ") #replace_dic의 value들을 공백+value로 바꾼다.

        text = text.strip()

    new_material.append(text)

new_material[100:105]
```

```python
len(new_material)
```

```python
# material의 (원자재_idx, 비율_idx, material)을 list안으로 저장함.. ex) 100 면, 면 100
check = 0
rows = []

for new in new_material:
    words = new.split()
    check_num = [w.isnumeric() for w in words]  # [T F T F] 형태로 저장됨.
    num_idx = [i for i, check in enumerate(check_num) if check]  # [0, 2] 숫자인(비율) 데이터의 인덱스를 저장

    if len(num_idx) == 0:
        material_idx = None
        check += 1

    elif check_num[0] == True and check_num[-1] == False:  # 숫자로 시작되며 맨뒤가 숫자가 아닌 경우
        material_idx = [n+1 for n in num_idx] # i + 1 원자재의 index를 저장함

    elif check_num[0] == False and check_num[-1] == True  :  # 원자재로 시작하며 맨뒤가 숫자로 끝나도록
        material_idx = [n-1 for n in num_idx]   # i - 1 원자재의 index를 저장

    else:
        material_idx = None
        check += 1

    row = (material_idx, num_idx, words)
    rows.append(row)

check
```

```python
# [{원자재:비율}, {원자재:비율, ...}...] 형태로 데이터를 저장하고 dataframe을 만든다.
materials = []
for mat_idxs, num_idxs, words in rows:
    material_map = {}
    if mat_idxs:
        for mat_idx, num_idx in zip(mat_idxs, num_idxs):
            material = words[mat_idx]
            pct = words[num_idx]
            material_map[material] = pct

    materials.append(material_map)

mdf = pd.DataFrame(materials)
mdf
```

```python
def single_price_category(p):
    if p < 30000:
        p_cat = 0
    elif 30000 <= p < 50000:
        p_cat = 1
    elif 50000 <= p < 70000:
        p_cat = 2
    elif 70000 <= p < 90000:
        p_cat = 3
    elif 90000 <= p < 110000:
        p_cat = 4
    elif 110000 <= p < 130000:
        p_cat = 5
    elif 130000 <= p < 180000:
        p_cat = 6
    else:  # 18만원 이상
        p_cat = 7
    return p_cat

df["price_category"] = df["price"].apply(single_price_category)
```
